###Pandas Series from Dictionary

```python
data_dict = {'A': 10, 'B': 20, 'C': 30, 'D': 40, 'E': 50} #dictionary object key-value pairs, #index is not predefined anymore, we define the index

series_from_dict = pd.Series(data_dict)
print(series_from_dict)

#dtype: int64 - integer values
```

```python
csv_data = "1,2,3,4,5" #string
csv_data.split(',') #sprit the string into a list
```

```python
series_from_list_split = pd.Series(csv_data.split(','))
print(series_from_list_split)
```
