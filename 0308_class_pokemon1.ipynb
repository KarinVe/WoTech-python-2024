```python
# required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #vizualization
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=10) #will give you top 10 rows
```

```python
pokemon_df.sample(5) #will give you random 5 rows, every time you run the code, you get different output
```

```python
# How many rows and columns are in this data?

rows, columns = pokemon_df.shape #stores the values behind rows and columns variable
print(f'Number of rows is {rows}, number of columns is {columns}')
```

```python
# How many unique Pokemon names are in this table?

unique_names = pokemon_df['Name'].nunique() # nunigue -> number of unique
unique_names
```

```python
# How many Pokemons have Attack greater than 100?

attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(len(attack_greater_than_100))
```

```python
# How many Pokemons have Defence greater than 100?
defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(len(defense_greater_than_100))
```

```python
# How many Pokemons have both greater than 100?
attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(len(attack_defense_greater_than_100))
```

```python
# Which pokemon has the biggest attack?

pokemon_df['Attack'].max() #the highest value of the attack
pokemon_df['Attack'].idxmax() #index of the highest value

highest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()] #iloc - index location
print('Pokemon with the highest Attack is:', highest_attack )
print(f'Pokemon with the highest Attack is: {highest_attack}')
```

```python
# Which pokemon has the lowest attack?

pokemon_df['Attack'].min() #the minimum value of the attack
pokemon_df['Attack'].idxmin() #index of the minimum value

lowest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmin()] #iloc - index location
print('Pokemon with the lowest Attack is:', lowest_attack )
print(f'Pokemon with the lowest Attack is: {lowest_attack}')
```

How many Pokémons are with 'Type 1' == Water as a % of total?

```python
type1_water_pokemons = pokemon_df[pokemon_df['Type 1'] =='Water'].value_counts()
type1_water_pokemons.sum()
```

```python
total_pokemons_type1 = pokemon_df['Type 1'].value_counts().sum()
total_pokemons_type1
```

```python
# find the percentage of total
percentage_of_water = type1_water_pokemons.sum() / total_pokemons_type1 * 100
print(f'There are {percentage_of_water:.1f}% of pokemons of Type1 == Water')
```

What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?

```python
max_value = pokemon_df['Speed'].max()
max_value
```

```python
min_value = pokemon_df['Speed'].min()
min_value
```

```python
speed_difference = max_value - min_value
speed_difference
```
